```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

### numpy Demo

```python
# 定义四叶草函数
t = np.arange(0,2*np.pi,0.01)
x = np.square(np.sin(t)) * np.cos(t)
y = np.square(np.cos(t)) * np.sin(t)
```

### matplotlib Demo

```python
plt.plot(x, y)
```

### pandas Demo

```python
# 一维数组创建series
arr = np.array([1, 2, 3, 4, 5, 6])
series = pd.Series(arr)
print(series)

# DataFrame
df = pd.DataFrame([[1, 2, 3], [4, 5, 6]])
print(df.count(axis = 0))
print(df.count(axis = 1))
```
